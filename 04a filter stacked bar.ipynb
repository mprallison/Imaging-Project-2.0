```python
import numpy as np
import pandas as pd
import bokeh
from bokeh.plotting import figure,show
from bokeh.io import output_notebook,output_file
from bokeh.models import CategoricalColorMapper

output_notebook()
```

```python
df = pd.read_csv('test files\detected_results.csv').head(100)
df = df.sort_values(by='ratio_red',ascending=False)
```

```python
#add bool_filter column
def cellFilter(df):
    a, b = (float(x) for x in input('red filter').split())
    c, d = (float(x) for x in input('green filter').split())
    e, f = (float(x) for x in input('blue filter').split())
    
    def filterDF(red,green,blue):
        if a <= red <= b or c <= green <= d or e <= blue <= f:
            return 'Y'
        else:
            return 'N'
        
    df['label'] = df[['ratio_red','ratio_green','ratio_blue']].apply(lambda x:filterDF(*x),axis=1)
    
    return df
```

```python
df = cellFilter(df)
```
